1. What is a `DataFrame` and a `Series`? What is the difference between the two?

A `Series` is a 1D array that holds any data type. This is simliar to a singular column in a SQL table

A `DataFrame` is a 2D collection of multiple `Series` combined together. It is like a spreadsheet or a singular SQL table.
